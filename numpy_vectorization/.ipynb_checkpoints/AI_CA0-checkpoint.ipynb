<div style="direction:rtl;font-size: 16px;text-align: justify;line-height: 1.6"> 
برای پر کردن داده های خالی روش های گوناگونی وجود دارد.یکی از این روش ها که در همین پروژه نیز از آن استفاده شده است جایگذاری میانگین هر ستون در سلول های خالی است.روش دیگر استفاده از میانه هر ستون است.باتوجه به نوع داده ای که به ما داده می شود ممکن است یکی از این روش ها بهتر عمل کند.برای مثال در صورتی که همه داده ها در رنج تقریبا یکسانی قرار داشته باشند و داده ی پرتی نداشته باشیم،استفاده از میانگین روش خوبی است چرا که در این صورت میانگین هر ستون می تواند نماینده ی خوبی برای داده های آن ستون باشد همچنین لازم نیست هزینه ای برای سورت داده ها بدهیم تا میانه را به دست آوریم در غیر این صورت ممکن است داده ای وجود داشته باشد که باعث شود میانگین کلی داده خیلی پایین یا خیلی بالا برود و باعث شود میانگین تخمین خیلی خوبی برای داده های خالی نباشد.در این صورت استفاده از میانه راه بهتری است چرا که داده ها سورت می شوند و داده ی میانی برای قسمت های خالی در نظر گرفته می شود.در این صورت نقش داده های پرت کم می شود.
    <br>
  یکی دیگر از روش ها،دسته بندی داده ها بسته براساس ویژگی های آن هاست یعنی براساس یک سری ویژگی داده  ها را دسته بندی کنیم تا داده های شبیه به هم در یک دسته قرار گیرند و برای هر سلول خالی میانگین دسته اش را جاگذاری کنیم نه میانگین کل داده ها.اگرچه این روش نسبت به میانگین هزینه ی بیشتری دارد ولی ممکن است درجاهایی که احتیاج به خطای کمتری داریم این راه بهتر عمل کند و داده های از رست رفته را با دقت بیشتری بازیابی کند.
</div>

```python
col = list(data)
col.remove('Clicked on Ad')
data[col] = data[col].fillna(value=data[col].mean())
data.head()
```

<div style="direction:rtl;font-size: 16px;text-align: justify;line-height: 1.6">
با گذاشتن شرط های مختلف بر روی داده می توان اطلاعات گوناگونی از آن استخراج کرد.در ادامه اطلاعاتی مانند تعداد کاربران زن و مرد،تعداد کاربرانی که روی آگهی کلیک کرده اند یا نکرده اند و تعداد کاربرانی که زن هستند و سن بیشتر از 21 سال دارند آمده است.
</div>

```python
genderCount = data['Gender'].value_counts()
print("Female: ",genderCount[0],"\nMale: ",genderCount[1])
```

```python
clickedCount = data['Clicked on Ad'].value_counts()
print("Has clicked: ",clickedCount[1],"\nHas not clicked: ",clickedCount[0])
```

```python
size = len(data[(data['Gender']==0) & (data['Age']>21)])
print("Genger = Female & Age > 21 : ",size)
```

<div style="direction:rtl;font-size: 16px;text-align: justify;line-height: 1.6">
همان طور که پیش تر گفته شد استفاده از vectorization باعث می شود سرعت انجام محاسبات بالاتر برود.در ادامه یک بار به کمک این روش و بار دیگر با استفاده از حلقه، میانگین سن کاربرانی که روی آگهی کلیک کرده اند و کاربرانی که روی آگهی کلیک نکرده اند به دست آمده و زمان مربوط به هر یک گزارش شده است.
</div>

```python
tic = time.time()
avgAge = data.groupby('Clicked on Ad')['Age'].mean()
toc = time.time()
print("time: ",toc-tic)
avgAge
```

```python
tic = time.time()
clickedAvg,clickedCount = 0,0
notClickedAvg,notClickedCount = 0,0
for i in range(0,len(data)):
    if(data.at[i,'Clicked on Ad']==1):
        clickedAvg += data.at[i,'Age']
        clickedCount +=1
    elif(data.at[i,'Clicked on Ad']==0):
        notClickedAvg += data.at[i,'Age']
        notClickedCount +=1

clickedAvg = clickedAvg/clickedCount
notClickedAvg = notClickedAvg/notClickedCount
toc = time.time()
print("time: ",toc-tic)
print("Clicked avg: ",clickedAvg,"\nNot clicked avg: ",notClickedAvg)
```

<div style="direction:rtl;font-size: 16px;text-align: justify;line-height: 1.6">
    همان طور که مشاهده می شود روش vectorization تقریبا ده برابر سریع تر عمل می کند.
</div>
<div style="direction:rtl;font-size: 16px;text-align: justify;line-height: 1.6">
    شکل زیر هیستوگرام مربوط به ستون های عددی است که شکل توزیع هر ستون را نشان میدهد.
</div>

```python
ax = data.drop(columns=['Unnamed: 0']).hist(figsize=(11,11))
for ax in ax.flatten():
    ax.set_xlabel(ax.get_title())
    ax.set_ylabel('count')
```

<div style="direction:rtl;font-size: 16px;text-align: justify;line-height: 1.6">
     در ادامه تنها از ستون هایی استفاده شده است که مقدار آن ها عددی است.بنابراین ستون های غیر عددی و ستون Gender دیگر در تحلیل داده ها نقشی ندارند.یکی از راه های بهبود داده ها برای مدل های یادگیری ماشین، نرمال سازی داده هاست.در این پروژه نرمال سازی برای هر ستون ویژگی،با کم کردن میانگین و تقسیم کردن بر انحراف معیار انجام شده است و داده نرمالایز شده در normalized_data ذخیره شده است.
</div>

```python
normalized_data = data.select_dtypes([np.number]).drop(columns=['Gender'])
col = list(normalized_data.columns)
col.remove('Unnamed: 0')
col.remove('Clicked on Ad')
normalized_data[col] = (data[col]-data[col].mean())/data[col].std()
normalized_data.head(20)
```

<div style="direction:rtl;font-size: 16px;text-align: justify;line-height: 1.6">
به کمک توابع mean و std برای هر دو حالتی که کاربر روی آگهی کلیک کرده و روی آگهی کلیک نکرده میانگین و انحراف معیار هر کدام از ویژگی ها به دست آمده و به کمک scipy.stats.norm.pdf تابع چگالی احتمال مربوط به هر ویژگی رسم شده است.
</div>

```python
mean = normalized_data.groupby('Clicked on Ad')[col].mean()
std = normalized_data.groupby('Clicked on Ad')[col].std()
```

```python
mean
```

```python
std
```

```python
fig,axs = plt.subplots(2,2,figsize=(11,11))
clickedRow = normalized_data.loc[normalized_data['Clicked on Ad']==1]
unclickedRow = normalized_data.loc[normalized_data['Clicked on Ad']==0]
for i,element in enumerate(col):
    x0 = unclickedRow[element]
    x1 = clickedRow[element]
    idx = [i//2,i%2]
    axs[idx[0],idx[1]].plot(x0,scipy.stats.norm.pdf(x0,mean[element][0],std[element][0]),'.',label= 'not clicked')
    axs[idx[0],idx[1]].plot(x1,scipy.stats.norm.pdf(x1,mean[element][1],std[element][1]),'.',label = 'clicked')
    axs[idx[0],idx[1]].set_title(element)
    axs[idx[0],idx[1]].set_xlabel('normalized '+element)
    axs[idx[0],idx[1]].set_ylabel('pdf')
    axs[idx[0],idx[1]].legend(loc="upper left")
```

<div style="direction:rtl;font-size: 16px;text-align: justify;line-height: 1.6">
برای انتخاب بهترین ویژگی باید نموداری را انتخاب کنیم که دو کلاس کلیک کرده و کلیک نکرده را بهتر تفکیک کرده باشد.به این منظور باید نموداری انتخاب شود که مساحت مشترک بین دو نمودار کمینه باشد تا احتمال خطا کمتر شود چرا که در این ناحیه داده ممکن است مربوط به کلاس کلیک نکرده یا کلیک کرده باشد.از شکل واضح است که دو نمودار مربوط به Age وAge Income مساحت مشترک بیشتری دارند.برای بررسی دقیق تر دو نمودار دیگر آن ها را روی هم می کشیم تا تصمیم گیری دقیق تری داشته باشیم.
</div>

```python
list_comp = ['Daily Internet Usage','Daily Time Spent on Site']
for element in list_comp:
    x0 = unclickedRow[element]
    x1 = clickedRow[element]
    plt.plot(x0,scipy.stats.norm.pdf(x0,mean[element][0],std[element][0]),'.',label= 'not clicked '+element)
    plt.plot(x1,scipy.stats.norm.pdf(x1,mean[element][1],std[element][1]),'.',label = 'clicked '+element)
    
plt.title('Daily Internet Usage vs Daily Time Spent on Site')    
plt.xlabel('normalized data')
plt.ylabel('pdf')
plt.legend(loc="upper right",bbox_to_anchor=(1.8, 0.7))
```

<div style="direction:rtl;font-size: 16px;text-align: justify;line-height: 1.6">
همان طور که از شکل پیداست نمودار نارنجی که مربوط به Daily Internet Usage است پایین تر از نمودار قرمز که مربوط به Daily Time Spent On Site است قرار دارد بنابراین ویژگی Daily Internet Usage دارای کمترین مساحت مشترک بین سایر ویژگی هاست و به همین جهت به عنوان بهترین ویژگی برای ورودی مدل انتخاب می شود.
</div>
<div style="direction:rtl;font-size: 16px;text-align: justify;line-height: 1.6">
    برای پیش بینی ستون هدف لازم است تا مرز کلاس ها را تعیین کنیم.همان طور که می دانیم برای محاسبه ی احتمال به کمک تابع چگالی احتمال از فرمول زیر استفاده می شود:
</div>
<img src="https://res.cloudinary.com/dawrzv8vt/image/upload/v1601306150/Screen_Shot_2020-09-28_at_6.44.33_PM_ecjv0l.png" style="width: 200px;"/>
<div style="direction:rtl;font-size: 16px;text-align: justify;line-height: 1.6;">
    بنابراین باید به ازای هر نقطه مشخص کنیم که تابع بالا به ازای کدام نمودار بیشتر می شود.برای هر کدام که بیشتر شد یعنی احتمال اینکه داده در آن  کلاس بوده بیشتر بوده و در نتیجه داده را در آن کلاس قرار می دهیم.از آن جا که عرض   در هر دو نمودار dx است در واقع باید ببینم در هر نقطه مقدار کدام نمودار بیشتر بوده است.با حل معادله زیر نقطه برخورد به دست می آید.  
</div> 
<br>   
\begin{align*}
\frac{1}{0.592058}e^{\frac{-1}{2 \times 0.592058^2}(x-0.764216)^2} - \frac{1}{0.723959
}e^{\frac{-1}{2 \times 0.723959^2}(x+0.753395)^2} =0
\end{align*}
<br>
<div style="direction:rtl;font-size: 16px;text-align: justify;line-height: 1.6;">
     این معادله دارای دو نقطه برخورد x1=0.025 و x2=7.63 است.x2 در رنج داده های ما نیست و آن را در نظر نمی گیریم.در نمودار زیر مرز x1 جهت جدا سازی دو کلاس کشیده شده است.
</div>


```python
selected = 'Daily Internet Usage'
x0 = unclickedRow[selected]
x1 = clickedRow[selected]
plt.plot(x0,scipy.stats.norm.pdf(x0,mean[selected][0],std[selected][0]),'.',label= 'not clicked')
plt.plot(x1,scipy.stats.norm.pdf(x1,mean[selected][1],std[selected][1]),'.',label = 'clicked')
plt.title(selected)
plt.xlabel('normalized '+selected)
plt.ylabel('pdf')
plt.legend(loc="upper left")
plt.axvline(x=0.025,ymin=0.025, ymax=0.98,color='r',linestyle='--')
```

<div style="direction:rtl;font-size: 16px;text-align: justify;line-height: 1.6">
از شکل پیداست که در نقاط کمتر از 0.025 نمودار نارنجی(کلیک) دارای مقدار بیشتری می باشد و در نقاط بیشتر از آن نمودار آبی(عدم کلیک) مقدار بیشتری می گیرد.بنابراین ستون هایی که مقدار ویژگی Daily Internet Usage(نرمالایز) آن کمتر یا مساوی 0.025 بود در کلاس کلیک کرده و بقیه در کلاس کلیک نکرده قرار می گیرند.
</div>

```python
nan_targets = normalized_data[normalized_data['Clicked on Ad'].isnull()].copy()
nan_targets['Clicked on Ad'] = np.where(nan_targets['Daily Internet Usage'] <=0.025,1.0,0.0)
nan_targets = nan_targets[['Clicked on Ad']]
nan_targets
```

</div>
