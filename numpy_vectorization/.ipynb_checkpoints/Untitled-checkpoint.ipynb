```python
import pandas as pd
import numpy as np
data = pd.read_csv('advertising_dataset.csv')
```

```python
data['Gender'] = np.where(data.Gender =='Male',1,0)
data.head()
```

```python
data.isna().sum()
```

```python
col = list(data)
col.remove('Clicked on Ad')
data[col] = data[col].fillna(value=data[col].mean())
data.head()
```

```python
gcount = data['Gender'].value_counts()
print("Female: ",gcount[0],"- Male: ",gcount[1])
```
